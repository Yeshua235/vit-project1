
# Comparing 10 Linear Dimensionality Reduction Methods (Latent=5, Observed=12)

We construct a **custom synthetic dataset** with **12 observed features** derived from **5 true latent dimensions** plus correlated and noise dimensions. We compare **ten widely used linear methods** for dimensionality reduction to 2, 4, 6, and 8 dimensions, then reconstruct the data and evaluate performance using simple matrix/vector metrics.

### Techniques compared (10)
1. **PCA** (Principal Component Analysis)
2. **IncrementalPCA** (memory-efficient PCA variant)
3. **Truncated SVD** (low-rank linear projection; PCA-like on centered data)
4. **Factor Analysis** (linear latent factor model)
5. **Sparse PCA** (PCA with sparsity in components)
6. **MiniBatch Sparse PCA** (mini-batch variant for large data)
7. **Dictionary Learning** (linear sparse coding)
8. **Gaussian Random Projection** (Johnson–Lindenstrauss linear map)

> We use **NumPy/SciPy** for metrics and **scikit‑learn** only for the dimensionality reduction algorithms.

### Metrics (concise definitions)
Let $X$ be the centered data and $\hat{X}$ its reconstruction from a lower-dimensional representation; residual $R = X - \hat{X}$. We report five metrics:

1. **Frobenius norm**: $\|R\|_F = (\sum_{i,j} R_{ij}^2)^{1/2}$ — overall residual energy.
2. **Relative Frobenius norm**: $\|R\|_F / \|X\|_F$ — scale-normalized residual.
3. **Spectral (operator 2-) norm**: largest singular value of $R$ — worst-case residual mode.
4. **Mean Squared Error (MSE)**: $\frac{1}{nm}\,\|R\|_F^2$ — average squared error per entry.
5. **Reconstruction $R^2$**: $1 - \frac{\|R\|_F^2}{\|X\|_F^2}$ — fraction of (centered) variance explained.


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm, pinv
from scipy.linalg import svdvals
from sklearn.decomposition import (
    PCA, IncrementalPCA, TruncatedSVD, FactorAnalysis, 
    SparsePCA, MiniBatchSparsePCA, DictionaryLearning
)
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection
```

```python
np.random.seed(42)

def generate_synthetic_data(n_samples=500, n_features=12, true_dim=5):
    '''
    12 observed features from 5 latent variables:
    - 9 features: linear mixtures (varying scales) + small measurement noise
    - 2 features: correlated combos of earlier mixture features + small noise
    - 1 feature: pure noise
    Returns centered X, mean, and latent sources S.
    '''
    # Latent sources (known true dimensions) with decreasing scales
    scales = np.array([3.0, 2.5, 2.0, 1.5, 1.0])
    S = np.random.randn(n_samples, true_dim) * scales

    # Mix into first 9 observed features
    n_mix = 9
    B = np.random.randn(true_dim, n_mix)
    Y = S @ B
    Y += 0.05 * np.random.randn(n_samples, n_mix)  # small measurement noise

    # Correlated features from mixtures
    f10 = 0.6 * Y[:, 0] - 0.4 * Y[:, 3] + 0.1 * np.random.randn(n_samples)
    f11 = -0.2 * Y[:, 5] + 0.5 * Y[:, 1] + 0.1 * np.random.randn(n_samples)

    # Pure noise feature
    f12 = 0.5 * np.random.randn(n_samples)

    X = np.column_stack([Y, f10, f11, f12])  # total 12 features

    # Global small noise across all features
    X += 0.02 * np.random.randn(n_samples, X.shape[1])

    mu = X.mean(axis=0)
    Xc = X - mu
    return Xc, mu, S
```

```python
# Metrics

def frobenius_norm(R):
    return norm(R, 'fro')

def relative_frobenius(R, Xc):
    return frobenius_norm(R) / (norm(Xc, 'fro') + 1e-12)

def spectral_norm(R):
    return svdvals(R)[0]

def mse(R):
    return np.mean(R**2)

def reconstruction_R2(R, Xc):
    return 1.0 - (frobenius_norm(R)**2) / (norm(Xc, 'fro')**2 + 1e-12)

metrics = {
    'Frobenius': frobenius_norm,
    'RelFro': relative_frobenius,
    'Spectral': spectral_norm,
    'MSE': mse,
    'R2': reconstruction_R2,
}
```

```python
# Unified reconstruction wrappers for each method

def reconstruct_with_pca(Xc, k):
    model = PCA(n_components=k, random_state=42)
    Z = model.fit_transform(Xc)
    Xhat = model.inverse_transform(Z)
    return Xhat

def reconstruct_with_ipca(Xc, k):
    model = IncrementalPCA(n_components=k)
    model.fit(Xc)  # single batch for comparability
    Z = model.transform(Xc)
    Xhat = model.inverse_transform(Z)
    return Xhat

def reconstruct_with_tsvd(Xc, k):
    model = TruncatedSVD(n_components=k, random_state=42)
    Z = model.fit_transform(Xc)
    if hasattr(model, 'inverse_transform'):
        Xhat = model.inverse_transform(Z)
    else:
        Xhat = Z @ model.components_
    return Xhat

def reconstruct_with_fa(Xc, k):
    model = FactorAnalysis(n_components=k, random_state=42, max_iter=1000)
    Z = model.fit_transform(Xc)
    if hasattr(model, 'inverse_transform'):
        Xhat = model.inverse_transform(Z)
    else:
        Xhat = Z @ model.components_
    return Xhat

def reconstruct_with_spca(Xc, k):
    model = SparsePCA(n_components=k, random_state=42, alpha=1e-3, max_iter=1000)
    Z = model.fit_transform(Xc)
    Xhat = Z @ model.components_
    return Xhat

def reconstruct_with_mbspca(Xc, k):
    model = MiniBatchSparsePCA(n_components=k, random_state=42, alpha=1e-3, max_iter=1000)
    Z = model.fit_transform(Xc)
    Xhat = Z @ model.components_
    return Xhat

def reconstruct_with_dictlearn(Xc, k):
    model = DictionaryLearning(n_components=k, random_state=42, max_iter=1000)
    Z = model.fit_transform(Xc)
    Xhat = Z @ model.components_
    return Xhat

def reconstruct_with_grp(Xc, k):
    model = GaussianRandomProjection(n_components=k, random_state=42)
    Z = model.fit_transform(Xc)
    W = model.components_            # (k, n_features)
    A = pinv(W.T)                    # (n_features, k)
    Xhat = Z @ A
    return Xhat

methods = {
    'PCA': reconstruct_with_pca,
    'IncrementalPCA': reconstruct_with_ipca,
    'TruncatedSVD': reconstruct_with_tsvd,
    'FactorAnalysis': reconstruct_with_fa,
    'SparsePCA': reconstruct_with_spca,
    'MiniBatchSparsePCA': reconstruct_with_mbspca,
    'DictionaryLearning': reconstruct_with_dictlearn,
    'GaussianRP': reconstruct_with_grp,
}
```

```python
# Generate data with latent_dim=5 and observed_dim=12
Xc, mu, S_true = generate_synthetic_data(n_samples=500, n_features=12, true_dim=5)
print(f'Centered data shape: {Xc.shape}, true latent dim: {S_true.shape[1]}')
print(f'Frobenius norm of centered data: {norm(Xc, "fro"):.4f}')
```

```python
# Evaluate each method at target dimensions

target_dims = [2, 4, 6, 8]
results = {d: [] for d in target_dims}

for d in target_dims:
    for name, recon_fn in methods.items():
        try:
            Xhat = recon_fn(Xc, d)
            R = Xc - Xhat
            row = {
                'Method': name,
                'Dim': d,
                'Frobenius': frobenius_norm(R),
                'RelFro': relative_frobenius(R, Xc),
                'Spectral': spectral_norm(R),
                'MSE': mse(R),
                'R2': reconstruction_R2(R, Xc),
            }
            results[d].append(row)
        except Exception as e:
            results[d].append({'Method': name, 'Dim': d, 'Error': str(e)})

```

```python
# Pretty print tables
def print_table(rows, headers):
    widths = [max(len(h), max((len(f"{r.get(h, '')}") for r in rows), default=0)) for h in headers]
    fmt_row = ' | '.join('{:'+str(w)+'}' for w in widths)
    sep = '-+-'.join('-'*w for w in widths)
    print(fmt_row.format(*headers))
    print(sep)
    for r in rows:
        vals = [r.get(h, '') for h in headers]
        print(fmt_row.format(*vals))

for d in target_dims:
    print(f"\n=== Results for target dimension {d} ===")
    rows = results[d]
    if any('Error' in r for r in rows):
        print_table(rows, ['Method', 'Dim', 'Error'])
    else:
        formatted = []
        for r in rows:
            formatted.append({
                'Method': r['Method'],
                'Dim': str(r['Dim']),
                'Frobenius': f"{r['Frobenius']:.6f}",
                'RelFro': f"{r['RelFro']:.6f}",
                'Spectral': f"{r['Spectral']:.6f}",
                'MSE': f"{r['MSE']:.6f}",
                'R2': f"{r['R2']:.6f}",
            })
        print_table(formatted, ['Method', 'Dim', 'Frobenius', 'RelFro', 'Spectral', 'MSE', 'R2'])
```

### Notes & Assumptions
- The dataset is **centered** before applying all methods; metrics are computed on centered $X$.
- **Random Projections** are not bijective; we reconstruct using the **Moore–Penrose pseudoinverse** for best least-squares fit.
- **SparsePCA / MiniBatchSparsePCA / DictionaryLearning** return codes and components; reconstruction uses `codes @ components_`.
- **IncrementalPCA** is evaluated in a single batch here for comparability; it matches PCA in result for centered data.
- The true latent dimension is **5**, so reductions with target dimension $\ge 5$ should (in noiseless settings) yield small residuals; noise and correlated features limit perfect recovery.


```python

# --- Metric trend plots (vs. dimensionality) ---
results_dict = results

# Flatten the results structure into simple rows
def flatten_results(results_dict):
    """
    Converts results={k_dim: [ {Method, Dim, <metrics>...} , ...], ...}
    into a flat list of rows while skipping entries with 'Error'.
    """
    rows = []
    for d, lst in results_dict.items():
        for r in lst:
            if 'Error' in r:
                # skip methods that failed at this dimensionality
                continue
            rows.append(r)
    return rows

flat = flatten_results(results)

# Determine the available methods and dims that actually have scores
metric_names = ['Frobenius', 'RelFro', 'Spectral', 'MSE', 'R2']
dim_list = sorted(results.keys())
methods_present = sorted({r['Method'] for r in flat})

# Build metric -> method -> values aligned to dim_list
metric_map = {
    m: {meth: [np.nan]*len(dim_list) for meth in methods_present}
    for m in metric_names
}
dim_index = {d: i for i, d in enumerate(dim_list)}
for r in flat:
    i = dim_index[r['Dim']]
    for m in metric_names:
        metric_map[m][r['Method']][i] = r[m]

# Plot styling
plt.style.use('seaborn-v0_8-whitegrid')
FIGSIZE = (9, 4)
MARKERS = ['o', 's', '^', 'D', 'v', 'h', 'p', 'X', '8', '*']

def maybe_log_scale(ax, metric):
    """Use a log scale for y when appropriate (norms and MSE often span orders of magnitude)."""
    if metric in ('Frobenius', 'Spectral', 'MSE'):
        # Only apply log scale if all plotted values are positive
        ymin, ymax = ax.get_ylim()
        if ymin > 0 and ymax / max(ymin, 1e-12) > 10:
            ax.set_yscale('log')

for m in metric_names:
    fig, ax = plt.subplots(figsize=FIGSIZE)
    for j, (meth, ys) in enumerate(metric_map[m].items()):
        ys = np.array(ys, dtype=float)
        # If all NaN, skip
        if np.all(np.isnan(ys)):
            continue
        # Plot only where values are present
        mask = ~np.isnan(ys)
        if not np.any(mask):
            continue
        ax.plot(
            np.array(dim_list)[mask],
            ys[mask],
            marker=MARKERS[j % len(MARKERS)],
            linewidth=2,
            markersize=6,
            label=meth
        )
    ax.set_xlabel('Reduced dimension (k)', fontsize=12)
    ax.set_ylabel('Reconstruction $R^2$' if m == 'R2' else m, fontsize=12)
    ax.set_title(f'{m} vs. dimensionality', fontsize=14)
    ax.grid(True, alpha=0.3)
    maybe_log_scale(ax, m)
    ax.legend(ncol=2, fontsize=9, frameon=True)
    fig.tight_layout()
    plt.show()
```

```python
# --- Summary ranking per metric at each target dimension ---

metric_names = ['Frobenius', 'RelFro', 'Spectral', 'MSE', 'R2']
higher_better = {'R2': True}  # Others default to lower-is-better

def rank_methods_at_dim(flat_rows, dim, metric, higher_is_better=False):
    """
    Returns a sorted list of (method, value) for a given dimension and metric.
    Skips NaNs and missing values. Sort order depends on higher_is_better.
    """
    pairs = []
    for r in flat_rows:
        if r.get('Dim') != dim:
            continue
        val = r.get(metric, None)
        if val is None:
            continue
        if isinstance(val, float) and np.isnan(val):
            continue
        pairs.append((r['Method'], float(val)))
    # Deduplicate by method (keep the first occurrence)
    seen = {}
    for meth, v in pairs:
        if meth not in seen:
            seen[meth] = v
    pairs = list(seen.items())
    pairs.sort(key=lambda x: x[1], reverse=higher_is_better)
    return pairs

def print_rankings(flat_rows, dims, metric_names, higher_better_map, top_n=None):
    """
    Prints rankings for each metric at each dimension.
    If top_n is None, prints all; otherwise prints only top_n entries.
    """
    for d in sorted(dims):
        print(f'\n=== Rankings at reduced dimension k={d} ===')
        for m in metric_names:
            hb = higher_better_map.get(m, False)
            ranked = rank_methods_at_dim(flat_rows, d, m, higher_is_better=hb)
            # Format header with direction
            direction = '(higher is better)' if hb else '(lower is better)'
            print(f'  {m} {direction}:')
            if len(ranked) == 0:
                print('    (no successful results)')
                continue
            show = ranked if top_n is None else ranked[:top_n]
            for rank, (meth, val) in enumerate(show, start=1):
                print(f'    {rank:>2}. {meth:28s} {val:.6f}')
            if top_n is not None and len(ranked) > top_n:
                best = ranked[0][1]
                worst = ranked[-1][1]
                print(f'       ... ({len(ranked)-top_n} more; best={best:.6f}, worst={worst:.6f})')

# Print full rankings for each metric and dimension
# print_rankings(flat, dim_list, metric_names, higher_better, top_n=None)
```

## Non-Linear D.R. Techniques

```python
from sklearn.decomposition import KernelPCA
from sklearn.manifold import TSNE, Isomap, LocallyLinearEmbedding, SpectralEmbedding
```

```python
# Helper: best linear reconstruction from embedding Z

def best_linear_reconstruction(Z, Xc):
    '''Least-squares decoder from embedding to original space.
    A = pinv(Z) @ Xc; Xhat = Z @ A.
    '''
    A = pinv(Z) @ Xc
    return Z @ A
```

```python
# Non-linear methods (5)

# def reconstruct_with_tsne(Xc, k):
#     tsne = TSNE(n_components=k, random_state=42, perplexity=30, learning_rate='auto', init='pca', method='barnes_hut')
#     Z = tsne.fit_transform(Xc)
#     return best_linear_reconstruction(Z, Xc)


def reconstruct_with_tsne(Xc, k):
    """
    t-SNE embedding with method chosen by dimensionality:
    - k <= 3  -> Barnes–Hut (fast, requires quad/oct-tree)
    - k >= 4  -> Exact t-SNE (O(n^2), slower but valid for higher dims)

    Reconstruction uses the global least-squares decoder:
        Xhat = Z @ pinv(Z) @ Xc
    """
    tsne_kwargs = dict(
        n_components=k,
        random_state=42,
        perplexity=30,
        init='pca',
        learning_rate='auto',
        metric='euclidean'
    )
    if k <= 3:
        tsne_kwargs.update(method='barnes_hut', angle=0.5)  # angle only for barnes_hut
    else:
        tsne_kwargs.update(method='exact')                  # no angle in exact mode

    tsne = TSNE(**tsne_kwargs)
    Z = tsne.fit_transform(Xc)
    # Least-squares reconstruction from embedding
    A = np.linalg.pinv(Z) @ Xc
    Xhat = Z @ A
    return Xhat

def reconstruct_with_isomap(Xc, k):
    iso = Isomap(n_neighbors=10, n_components=k)
    Z = iso.fit_transform(Xc)
    return best_linear_reconstruction(Z, Xc)

def reconstruct_with_lle(Xc, k):
    lle = LocallyLinearEmbedding(n_neighbors=10, n_components=k, method='standard', random_state=42)
    Z = lle.fit_transform(Xc)
    return best_linear_reconstruction(Z, Xc)

def reconstruct_with_kpca(Xc, k):
    gamma = 1.0 / Xc.shape[1]
    kpca = KernelPCA(n_components=k, kernel='rbf', gamma=gamma, fit_inverse_transform=True, alpha=1e-3)
    Z = kpca.fit_transform(Xc)
    try:
        Xhat = kpca.inverse_transform(Z)
    except Exception:
        Xhat = best_linear_reconstruction(Z, Xc)
    return Xhat

def reconstruct_with_spectral(Xc, k):
    se = SpectralEmbedding(n_components=k, n_neighbors=10, random_state=42)
    Z = se.fit_transform(Xc)
    return best_linear_reconstruction(Z, Xc)

nonlinear_methods = {
    'tSNE': reconstruct_with_tsne,
    'Isomap': reconstruct_with_isomap,
    'LLE': reconstruct_with_lle,
    'KernelPCA': reconstruct_with_kpca,
    'SpectralEmbedding': reconstruct_with_spectral,
}
```

```python
# Utilities for evaluation & tabular printing

def evaluate_methods(Xc, methods_dict, target_dims):
    results = {d: [] for d in target_dims}
    for d in target_dims:
        for name, fn in methods_dict.items():
            try:
                Xhat = fn(Xc, d)
                R = Xc - Xhat
                row = {
                    'Method': name,
                    'Dim': d,
                    'Frobenius': frobenius_norm(R),
                    'RelFro': relative_frobenius(R, Xc),
                    'Spectral': spectral_norm(R),
                    'MSE': mse(R),
                    'R2': reconstruction_R2(R, Xc),
                }
                results[d].append(row)
            except Exception as e:
                results[d].append({'Method': name, 'Dim': d, 'Error': str(e)})
    return results

def print_table(rows, headers):
    widths = [max(len(h), max((len(f"{r.get(h, '')}") for r in rows), default=0)) for h in headers]
    fmt_row = ' | '.join('{:'+str(w)+'}' for w in widths)
    sep = '-+-'.join('-'*w for w in widths)
    print(fmt_row.format(*headers))
    print(sep)
    for r in rows:
        vals = [r.get(h, '') for h in headers]
        print(fmt_row.format(*vals))
```

```python
# Evaluate linear pool and select top-3 by average R2 across dims

target_dims = [2, 4, 6, 8]
linear_results = results

# Aggregate average R2 per method
avg_R2 = {}
for d, rows in linear_results.items():
    for r in rows:
        if 'Error' in r:
            continue
        avg_R2.setdefault(r['Method'], []).append(r['R2'])

avg_R2 = {m: (np.mean(v) if len(v) > 0 else -np.inf) for m, v in avg_R2.items()}
# Sort descending by average R2
top3_linear = sorted(avg_R2.items(), key=lambda x: x[1], reverse=True)[:3]
print("\nTop-3 linear methods by average R2 across dims:")
for rank, (meth, score) in enumerate(top3_linear, start=1):
    print(f"  {rank}. {meth:28s} R2_avg={score:.6f}")

selected_linear = {meth: methods[meth] for meth, _ in top3_linear}
```

```python
# Evaluate non-linear + selected linear baselines

combined_methods = {}
combined_methods.update(nonlinear_methods)
combined_methods.update(selected_linear)

combined_results = evaluate_methods(Xc, combined_methods, target_dims)

# Print tables per target dimension
for d in target_dims:
    rows = combined_results[d]
    print(f"\n=== Results (non-linear + top-3 linear) for k={d} ===")
    if any('Error' in r for r in rows):
        print_table(rows, ['Method', 'Dim', 'Error'])
    else:
        formatted = []
        for r in rows:
            formatted.append({
                'Method': r['Method'],
                'Dim': str(r['Dim']),
                'Frobenius': f"{r['Frobenius']:.6f}",
                'RelFro': f"{r['RelFro']:.6f}",
                'Spectral': f"{r['Spectral']:.6f}",
                'MSE': f"{r['MSE']:.6f}",
                'R2': f"{r['R2']:.6f}",
            })
        print_table(formatted, ['Method', 'Dim', 'Frobenius', 'RelFro', 'Spectral', 'MSE', 'R2'])
```

```python
# Metric trend plots & summary rankings

import matplotlib.pyplot as plt

# Flatten combined_results
flat = []
for d, lst in combined_results.items():
    for r in lst:
        if 'Error' in r:
            continue
        flat.append(r)

metric_names = ['Frobenius', 'RelFro', 'Spectral', 'MSE', 'R2']
dim_list = sorted(combined_results.keys())
methods_present = sorted({r['Method'] for r in flat})

metric_map = {m: {meth: [np.nan]*len(dim_list) for meth in methods_present} for m in metric_names}
index_of_dim = {d: i for i, d in enumerate(dim_list)}
for r in flat:
    i = index_of_dim[r['Dim']]
    for m in metric_names:
        metric_map[m][r['Method']][i] = r[m]

plt.style.use('seaborn-v0_8-whitegrid')
FIGSIZE = (10, 5)
MARKERS = ['o', 's', '^', 'D', 'v', 'h', 'p', 'X', '8', '*']

for m in metric_names:
    fig, ax = plt.subplots(figsize=FIGSIZE)
    for j, (meth, ys) in enumerate(metric_map[m].items()):
        ys = np.array(ys, dtype=float)
        mask = ~np.isnan(ys)
        if not np.any(mask):
            continue
        ax.plot(np.array(dim_list)[mask], ys[mask], marker=MARKERS[j % len(MARKERS)], linewidth=2, markersize=6, label=meth)
    ax.set_xlabel('Reduced dimension (k)', fontsize=12)
    ax.set_ylabel('Reconstruction $R^2$' if m == 'R2' else m, fontsize=12)
    ax.set_title(f'{m} vs. dimensionality', fontsize=14)
    ax.grid(True, alpha=0.3)
    if m in ('Frobenius', 'Spectral', 'MSE'):
        ymin, ymax = ax.get_ylim()
        if ymin > 0 and ymax / max(ymin, 1e-12) > 10:
            ax.set_yscale('log')
    ax.legend(ncol=2, fontsize=9, frameon=True)
    fig.tight_layout()
    plt.show()
```

```python
# Summary ranking per metric at each dimension

higher_better = {'R2': True}

def rank_methods_at_dim(flat_rows, dim, metric, higher_is_better=False):
    pairs = []
    for r in flat_rows:
        if r.get('Dim') != dim:
            continue
        val = r.get(metric, None)
        if val is None or (isinstance(val, float) and np.isnan(val)):
            continue
        pairs.append((r['Method'], float(val)))
    seen = {}
    for meth, v in pairs:
        if meth not in seen:
            seen[meth] = v
    pairs = list(seen.items())
    pairs.sort(key=lambda x: x[1], reverse=higher_is_better)
    return pairs

for d in dim_list:
    print(f"\n=== Rankings at reduced dimension k={d} ===")
    for m in metric_names:
        hb = higher_better.get(m, False)
        ranked = rank_methods_at_dim(flat, d, m, higher_is_better=hb)
        direction = '(higher is better)' if hb else '(lower is better)'
        print(f'  {m} {direction}:')
        if len(ranked) == 0:
            print('    (no successful results)')
            continue
        for rank, (meth, val) in enumerate(ranked, start=1):
            print(f'    {rank:>2}. {meth:28s} {val:.6f}')
```
